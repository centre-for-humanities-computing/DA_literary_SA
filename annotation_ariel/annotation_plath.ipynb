# Annotation instructions:

Score each verse on valence, from negative to positive on the scale 0-10. 
Try not to think about the context too much, that is, try to evaluate each verse individually, independently from what preceded it.
Some verses may not be complete senteces, since the sentences don not follow normal syntax while a sentence often includes several verses. Try not to worry about it, even if the words makes no sense. It is intented to be splitted up like this.

When you run the below cell, the text, verse per verse will show up in the terminal above. Enter a number and press ENTER. Then the next verse will show up.

Change the range before you start. The data contains 1914 verses in total. You just have to continue from where you left, if you need a break or don't have time to do them all. 
To exit the annotation, you can enter 's'. The results will be saved up to that point (every 10). Then just check the len of the dictionary at the bottom, set the annotation to start from this number 
and please remember to revise the pathnames above (so that you make a prv_savings_2.json, etc)

```python
scores = []
json_scores = {}

# Open the txt file in write mode
with open(out_path_json, "w") as out_txt:
    for i in range(len(data)):
        s = data[i]
        print(str(i) + "\t" + s)
        score = input(s)

        # Check for STOP
        if score.lower() == 's':
            break

        scores.append([score, s])

        # add to json dictionary
        json_scores[i] = [score, s]

        if len(scores) % 10 == 0:
            for j in range(len(scores)):
                score = scores[j]
                # Writing to txt
                out_txt.write(str(j) + ',' + '\t' + score[0] + ',' + score[1] + "\n")

            # Writing to JSON
            with open(out_path_json, "a") as json_file:
                json.dump(json_scores, json_file)

# Put remaining scores in the full json file
with open(out_path_json_full, "a") as json_file_all:
    json.dump(json_scores, json_file_all)



```

```python
with open(out_path_json_full, 'r') as file:
    annotated = json.load(file)

print(len(annotated))

pd.DataFrame.from_dict(annotated, orient='index', columns=['SCORE','SENTENCE'])
```
